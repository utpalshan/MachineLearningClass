Task-1:1.1 - Problem Statement

Write a Python Program to implement your own myreduce() function which works exactly like
Python's built-in function reduce()

```python
#Solution for Task-1:1.1:

#myreduce function to reduce and produce a single result for addition and multiplication

def myreduce(inp_func, input_itegers):
    output = input_itegers[0]
    for i in input_itegers[1:]:
        output = inp_func(output, i)    
    return output

#function to perform addition of 2 numbers
def addition(x,y):
    return x+y

#function to perform multiplication of 2 numbers
def multiplication(x,y):
    return x*y

def main_func():
    import time
    
    valid_functions = ['addition','multiplication']
    print("\nPlease enter a function one amongst:",valid_functions)
    time.sleep(1)
    
    input_function = input("Please enter function you would like myreduce to perform: ")
    inp_func = input_function.lower().strip()
    
    if inp_func in valid_functions:
        input_string = input("Please enter the list of input integers seperated by comma: ")
        input_list = input_string.split(",")
        length_input_list = len(input_list)
        
        for i in range(0,length_input_list): 
            input_list[i] = int(input_list[i])
        
        if inp_func == valid_functions[0]:
            result = myreduce(addition, input_list)
        else:
            result = myreduce(multiplication, input_list)
        print ("\nResult of input integer list using myreduce function is:",result)
    else:
        print("\nEntered input function for my reduce to perform is:",input_function,".Please enter valid function to continue.")
    
    time.sleep(1)
    option = input("Would you like to continue (Yes/No)?")
    if option.lower().strip() in ('yes', 'y'):
        main_func()
    else:
        print("Quitting the program.")

main_func()
```

```python
#Solution for Task-1:1.2:

# Custom filter function
def myfilter(inp_func, input_var):
    output = []
    for i in input_var:
        if inp_func(i):
            output.append(i)   
    
    return output

# function that filters vowels
def FilterVowels(input_var):
    vowels = ['a', 'e', 'i', 'o', 'u','A','E','I','O','U']

    if(input_var in vowels):
        return True
    else:
        return False

# function that filters positive numbers
def FilterPositive(input_var):
    if input_var > 0:
        return True
    else:
        return False

# function that filters even numbers
def FilterEven(input_var):
    if input_var % 2 == 0:
        return True
    else:
        return False

def main_func():
    import time
    
    valid_functions = ['filtervowels','filterpositive','filtereven']
    print("\nPlease enter a function one amongst:",valid_functions)
    time.sleep(1)
    
    input_function = input("Please enter function you would like myfilter to perform: ")
    inp_func = input_function.lower().strip()
    
    if inp_func in valid_functions:
        if inp_func == valid_functions[0]:
            input_string = input("Please enter the list of alphabets seperated by comma: ")
        else:
            input_string = input("Please enter the list of input integers seperated by comma: ")

        input_list = input_string.split(",")
        length_input_list = len(input_list)
        
        if inp_func == valid_functions[0]:
            result = myfilter(FilterVowels, input_list)
        else:
            for i in range(0,length_input_list): 
                input_list[i] = int(input_list[i])
            
            if inp_func == valid_functions[1]:
                result = myfilter(FilterPositive, input_list) 
            else:
                result = myfilter(FilterEven, input_list)
        
        print ("\nResult of myfilter function on the list of input passed is:",set(result))
    else:
        print("\nEntered input function for myfilter to perform is:",input_function)
    
    time.sleep(1)
    option = input("Would you like to continue (Yes/No)?")
    if option.lower().strip() in ('yes', 'y'):
        main_func()
    else:
        print("Quitting the program.")

main_func()
```

Task-1:3- Problem Statement

Implement a function longestWord() that takes a list of words and returns the longest one.

```python
#Solution for Task-1:3
 
input_strings = input("Please enter list of words delimited by space: ")
list_of_words = input_strings.split(" ") 
list_of_words.sort(key=len,reverse=True)

total_words = len(list_of_words)
length_of_longest_word = len(list_of_words[0])
output_result=[]

print("Length of longest word is:",length_of_longest_word)
for i in range(0,total_words):
    if (len(list_of_words[i]) == length_of_longest_word):
        output_result.append(list_of_words[i])
    
print("Longest word from key-in set of words is:", set(output_result))
```

Task-2:1.1 - Problem Statement

Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.

area = (s*(s-a)*(s-b)*(s-c)) ** 0.5

Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.

```python
#Solution for Task-2:1.1
import math

class Area_of_Triangle:
    def __init__(classT, a, b, c):
        classT.a = a
        classT.b = b
        classT.c = c
        classT.s = (a + b + c) * 0.5
    
    def Calc_Area_of_Triangle(classT):
        return (math.sqrt(classT.s * (classT.s - classT.a) * (classT.s - classT.b) * (classT.s - classT.c)))
    
def main_func():
    import time

    a  = int(input("Please enter side a: "))
    b  = int(input("Please enter side b: "))
    c  = int(input("Please enter side c: "))

    #Check for traingle inequality
    if (a+b <= c):
        print("\nEntered sides do not satisfy the triangle inequality.\nThe sum of the lengths of sides a, b must be greater than the length of the remaining side c")
    else:
        area_of_traingle_T = Area_of_Triangle(a,b,c)
        print("\nArea of Triangle is: ", area_of_traingle_T.Calc_Area_of_Triangle())

    time.sleep(1)
    option = input("\nWould you like to continue (Yes/No)?")
    if option.lower().strip() in ('yes', 'y'):
        main_func()
    else:
        print("Quitting the program.")

main_func()
```

Task-2:1.2 - Problem Statement

Write a function filter_long_words() that takes a list of words and an integer n and returns the list
of words that are longer than n.

```python
#Solution for Task-2:1.2

input_strings = input("Please enter list of words delimited by space: ")
list_of_words = input_strings.split(" ")
input_size = int(input("Please enter integer n: "))

no_of_input_words = len(list_of_words)
output_string_list = []

for i in range(no_of_input_words):
        if len(list_of_words[i]) > input_size:
            output_string_list.append(list_of_words[i])

print("\nWords with length greater than", input_size, "are per below:\n",set(output_string_list))
```
